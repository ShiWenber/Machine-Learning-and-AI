```python
import torch

class Net(torch.nn.Module):
    def forward(self, x):
        z1 = 2*x[1] + x[2]
        z2 = x[1] * 3*x[3]
        z3 = -x[3]

        u1 = torch.sin(z1)
        u2 = 2*x[3] + z2
        u3 = 2*z1 + z3

        v1 = u1 - u3
        v2 = torch.sin(-u2)
        v3 = u1 * u3

        # y = torch.tensor([0.0, 0.0])
        # y[0] = torch.pow(v1, 2) + torch.pow(v3, 3)
        # y[1] = v2 * v3
        y=v2 * v3
        return y



class Net2(torch.nn.Module):
    def forward(self, x):
        z1 = 2*x[1] + x[2]
        z2 = x[1] * 3*x[3]
        z3 = -x[3]

        u1 = torch.sin(z1)
        u2 = 2*x[3] + z2
        u3 = 2*z1 + z3

        v1 = u1 - u3
        v2 = torch.sin(-u2)
        v3 = u1 * u3

        # y = torch.tensor([0.0, 0.0])
        # y[0] = torch.pow(v1, 2) + torch.pow(v3, 3)
        # y[1] = v2 * v3
        y= torch.pow(v1, 2) + torch.pow(v2, 3)
        return y

```

```python
x = torch.tensor([0, 2, 1, 3], dtype=float,requires_grad=True)
model = Net()
y = model(x)
y.backward()
print(x.grad)
print( x.grad.equal(torch.tensor([0.0, 25.7484, 0.0614, 23.6467], dtype=torch.float64)))

# 获取需要更新的参数
params_to_update = []
for name, param in model.named_parameters():
    if param.requires_grad == True:
        params_to_update.append(param)
        print(name)
        
print(params_to_update)

x2 = torch.tensor([0, 2, 1,3], dtype=float,requires_grad=True)
model = Net2()
y = model(x2)
y.backward()
print(x2.grad)
# 判断x2.grad 是否正确，是否等于torch.tensor([0.0, 45.2487, 27.3204, -24.2664])
```

```python
# numpy
import numpy as np

class Net3():
    def forward(self, x):
        z1 = 2*x[1] + x[2]
        z2 = x[1] * 3*x[3]
        z3 = -x[3]

        u1 = np.sin(z1)
        u2 = 2*x[3] + z2
        u3 = 2*z1 + z3

        v1 = u1 - u3
        v2 = np.sin(-u2)
        v3 = u1 * u3

        # y = np.tensor([0.0, 0.0])
        # y[0] = np.pow(v1, 2) + torch.pow(v3, 3)
        # y[1] = v2 * v3
        y=v2 * v3
        return y

    def backward(self, x):
        z1 = 2*x[1] + x[2]
        z2 = x[1] * 3*x[3]
        z3 = -x[3]

        u1 = np.sin(z1)
        u2 = 2*x[3] + z2
        u3 = 2*z1 + z3

        v1 = u1 - u3
        v2 = np.sin(-u2)
        v3 = u1 * u3

        # y = np.tensor([0.0, 0.0])
        # y[0] = np.pow(v1, 2) + torch.pow(v3, 3)
        # y[1] = v2 * v3
        # y=v2 * v3

        y1 = np.zeros(4) 
        y1[0] = 2 * np.cos(z1) * 2 * v1 + (-8*v1) + 3*x[3] *(-np.cos(u2)) * 3 * np.power(v2, 2)
        y1[1] = np.cos(z1) * 2 * v1 + (- 4 * v1)
        y1[2] = -3*x[1]*np.cos(-u2) * 3 * np.power(v2, 2) - 2 * np.cos(-u2) * 3 * np.power(v2, 2) + 2 * v1

        y2 = np.zeros(4)
        y2[0] = 2 * np.cos(z1) * u3 * v2 + 4 * u1 * v2 + 3 * x[3] * v3 * (-np.cos(-u2))
        y2[1] = np.cos(z1) * u3 * v2 + 2 * u1 * v2
        y2[2] = -3 * x[1] * np.cos(-u2) * v3 - 2 * np.cos(-u2) * v3 - u1 * v2
        
        # 输出所有变量

        return y1, y2

```

```python
# 1
ls = [0,1,1,1]

## 对比torch自动求导和手动求导结果
x2 = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net2()
y = model(x2)
y.backward()
print(x2.grad)

x = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net()
y = model(x)
y.backward()
print(x.grad)

x3 = np.array(ls, dtype=float)
model = Net3()
y1, y2 = model.backward(x3)
print(y1)
print(y2)

# 2
ls = [0,2,1,3]

x2 = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net2()
y = model(x2)
y.backward()
print(x2.grad)

x = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net()
y = model(x)
y.backward()
print(x.grad)

## 对比torch自动求导和手动求导结果
x3 = np.array(ls, dtype=float)
model = Net3()
y1, y2 = model.backward(x3)
print(y1)
print(y2)

# 3
ls = [0,-1,-1,-3]

x2 = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net2()
y = model(x2)
y.backward()
print(x2.grad)

x = torch.tensor(ls, dtype=float,requires_grad=True)
model = Net()
y = model(x)
y.backward()
print(x.grad)

x3 = np.array(ls, dtype=float)
model = Net3()
y1, y2 = model.backward(x3)
print(y1)
print(y2)
```
